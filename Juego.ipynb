```python
import cv2
import random
import SeguimientoManos as sm  # Clase manos
import os
import imutils
```

```python
# Declaracion de variables 1 jugador
fs = False      # Bandera de haber presionado s
fu = False      # Bandera up
fd = False      # Bandera down
fj = False      # Bandera juego
fr = False      # Bandera reset
fgia = False    # Bandera gana IA
fgus = False    # Bandera gana Usuario
femp = False    # Bandera empate
fder = False    # Bandera derecha
fizq = False    # Bandera izquierda
conteo = 0
conia=0
conp1=0
```

```python
# Declaracion de variables 2 jugador
fini = False    # Bandera Iniciar
fu2 = False      # Bandera up
fd2 = False      # Bandera down
fpunt = False    # Bandera puntaje
ganoDerecha = False # Bandera de gano derecha
ganoIzquierda = False # bandera de gano izquierda
empate = False;  # bandera de empate
reset2 = False      # bandera para resetear
contp1 = 0    
contp2 = 0    
conteo2 = 0   
pizq = 0    
pder = 0   
```

```python
# Accedemos a la carpeta
path = 'Imagenes'
images = []
clases = []
lista = os.listdir(path)

# Leemos los rostros del DB
for lis in lista:
    # Leemos las imagenes de los rostros
    imgdb = cv2.imread(f'{path}/{lis}')
    # Almacenamos imagen
    images.append(imgdb)
    # Almacenamos nombre
    clases.append(os.path.splitext(lis)[0])
```

```python
def reset():
  
    fs = False      # Bandera de haber presionado s
    fu = False      # Bandera up
    fd = False      # Bandera down
    fj = False      # Bandera juego
    fr = False      # Bandera reset
    fgia = False    # Bandera gana IA
    fgus = False    # Bandera gana Usuario
    femp = False    # Bandera empate
    fder = False    # Bandera derecha
    fizq = False    # Bandera izquierda
    conteo = 0
    #Variables Jugador 2
    fini=False
    fu2=False
    fd2=False
    fpunt=False
    conteo2=0
    pizq=0
    pder=0
    ganoDerecha = False # Bandera de gano derecha
    ganoIzquierda = False # bandera de gano izquierda
    empate = False;  # bandera de empate
    reset2 = False      # bandera para resetear
    return fs,fu,fd,fj,fr,fgia,fgus,femp,fder,fizq,conteo,fini,fu2,fd2,fpunt,conteo2,pizq,pder,ganoDerecha,ganoIzquierda,empate, reset2
def ganadorIA():
    # GANA IA
    print('GANA LA IA')
    # Bandera Ganador
    fgia = True
    fgus = False
    femp = False
    fr = True
    return fgia,fgus,femp,fr
def empate():
    # EMPATE
    print('EMPATE')
    # Bandera Ganador
    fgia = False
    fgus = False
    femp = True
    fr = True
    return fgia,fgus,femp,fr
def ganadorUsuario():
    # GANA USUARIO
    print('GANA EL HUMANO')
    # Bandera Ganador
    fgia = False
    fgus = True
    femp = False
    fr = True
    return fgia,fgus,femp,fr
def mostrarIMG6():
    # Mostramos
    gan = images[6]
    alig, anig, c = gan.shape
    # Mostramos imagen
    frame[70: 70 + alig, 180: 180 + anig] = gan
def mostrarIMG7():
    # Mostramos
    gan = images[7]
    alig, anig, c = gan.shape
    # Mostramos imagen
    frame[70: 70 + alig, 180: 180 + anig] = gan
def mostrarIMG8():
    # Mostramos
    gan = images[8]
    alig, anig, c = gan.shape
    # Mostramos imagen
    frame[70: 70 + alig, 180: 180 + anig] = gan
    
    
```

```python
# Lectura de la camara
cap = cv2.VideoCapture(0)

# Declaramos el detector
detector = sm.detectormanos(Confdeteccion=0.9)

# Empezamos
while True:
    # Lectura de la videocaptura
    ret, frame = cap.read()

    # Leemos teclado
    t = cv2.waitKey(1)

    # Extraemos la mitad del frame
    al, an, c = frame.shape
    cx = int(an/2)
    cy = int(al/2)

    # Espejo
    frame = cv2.flip(frame,1)

    # Encontramos las manos
    frame = detector.encontrarmanos(frame, dibujar=True)
    # Posiciones mano 1
    lista1, bbox1, jug = detector.encontrarposicion(frame, ManoNum=0, dibujar=True, color = [0,255,0])

    # 1 Jugador
    if jug == 1:
        # Dividimos pantalla
        cv2.line(frame, (cx,0), (cx,480), (0,255,0), 2)

        # Mostramos jugadores
        #cv2.rectangle(frame, (245, 25), (380, 60), (0, 0, 0), -1)
        #cv2.putText(frame, '1 JUGADOR', (250, 50), cv2.FONT_HERSHEY_SIMPLEX, 0.71,(0, 255, 0), 2)


        #Puntaje
        cv2.putText(frame,str(conia),(288,90),cv2.FONT_HERSHEY_SIMPLEX,1.5,(0.255,0),3)
        cv2.putText(frame,str(conp1),(330,90),cv2.FONT_HERSHEY_SIMPLEX,1.5,(0.255,0),3)
 
        if fs==False:
            # Mensaje inicio
            cv2.rectangle(frame, (145, 425), (465, 460), (0, 0, 0), -1)
            cv2.putText(frame, 'PRESIONA S PARA EMPEZAR', (150, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.71, (0, 255, 0), 2)
        
        
            

        # Presiona S
        if t == 83 or t == 115 or fs == True:
            # Cambiamos bandera 's'
            fs = True
            # Obtenemos posicion de la mano
            if len(lista1) != 0:
                # Extraemos las coordenadas del dedo corazon
                x1, y1 = lista1[9][1:]

                #print(conteo)
                # Conteo de juego
                if conteo <= 2:
                    # Leemos la imagen inicial
                  
                    img = images[conteo]
                    # Redimensionamos
                    img = imutils.resize(img, width=240, height=240)
                    ali, ani, c = img.shape

                    # Definimos en que parte de la pantalla esta
                    # Izquierda
                    if x1 < cx:
                        # Banderas Lado
                        fizq = True
                        fder = False
                        # Mostramos imagen
                        frame[130: 130 + ali, 350: 350 + ani] = img

                        # Empezamos conteo
                        # Superamos el Umbral
                        if y1 < cy - 40 and fd == False:
                            fu = True
                            cv2.circle(frame, (cx, cy), 5, (255, 255, 0), -1)
                      

                        # Bajamos del Umbral con Flag
                        elif y1 > cy - 40 and fu == True:
                            conteo = conteo + 1
                            fu = False

                    # Derecha
                    elif x1 > cx:
                        # Banderas Lado
                        fder = True
                        fizq = False
                        # Mostramos la imagen
                        frame[130: 130 + ali, 30: 30 + ani] = img

                        # Empezamos conteo
                        # Superamos el Umbral
                        if y1 < cy - 40 and fd == False:
                            fu = True
                            cv2.circle(frame, (cx,cy), 5, (255,255,0), -1)
                            

                        # Bajamos del Umbral con Flag
                        elif y1 > cy - 40 and fu == True:
                            conteo = conteo + 1
                            fu = False
                # Jugamos
                elif conteo == 3:
                    # Si no hemos jugado jugamos
                    if fj == False and fr == False:
                        # Elegimos piedra papel o tijera
                        juego = random.randint(3,5)
                        fj = True

                    # Izquierda
                    if fizq == True and fder == False:
                        # Mostramos
                        img = images[juego]
                        # Redimensionamos
                        img = imutils.resize(img, width=240, height=240)
                        ali, ani, c = img.shape
                        # Mostramos imagen
                        frame[130: 130 + ali, 350: 350 + ani] = img
                        

                        # Si ya jugamos
                        if fj == True and fr == False:
                            # Extraemos valores de interes
                            # Punta DI
                            x2, y2 = lista1[8][1:]
                            # Punta DC
                            x3, y3 = lista1[12][1:]
                            # Punta DA
                            x4, y4 = lista1[16][1:]

                            # Falange DI
                            x22, y22 = lista1[6][1:]
                            # Falange DC
                            x33, y33 = lista1[10][1:]
                            # Falange DA
                            x44, y44 = lista1[14][1:]

                            # Condiciones de posicion
                            # Piedra
                            if x2 < x22 and x3 < x33 and x4 < x44:
                                # IA PAPEL
                                if juego == 3:
                                    fgia,fgus,femp,fr=ganadorIA()
                                    conia=conia+1

                                # IA PIEDRA
                                elif juego == 4:
                                    fgia,fgus,femp,fr=empate()
                                    
                                # IA TIJERA
                                elif juego == 5:
                                    fgia,fgus,femp,fr=ganadorUsuario()
                                    conp1=conp1+1



                            # Papel
                            elif x2 > x22 and x3 > x33 and x4 > x44:
                                # IA PAPEL
                                if juego == 3:
                                    fgia,fgus,femp,fr=empate()
                                # IA PIEDRA
                                elif juego == 4:
                                    fgia,fgus,femp,fr=ganadorUsuario()
                                    conp1=conp1+1
                                #IA Tijera
                                elif juego == 5:
                                    fgia,fgus,femp,fr=ganadorIA()
                                    conia=conia+1

                            # Tijera
                            elif x2 > x22 and x3 > x33 and x4 < x44:
                                # IA PAPEL
                                if juego == 3:
                                    fgia,fgus,femp,fr=ganadorUsuario()
                                    conp1=conp1+1

                                # IA PIEDRA
                                elif juego == 4:
                                    fgia,fgus,femp,fr=ganadorIA()
                                    conia=conia+1
                               #IA Tijeras
                                elif juego == 5:
                                    fgia,fgus,femp,fr=empate()

                        # Mostramos ganador
                        # IA
                        if fgia == True:
                            # Mostramos
                            mostrarIMG6()

                        # USUARIO
                        elif fgus == True:
                            mostrarIMG7()

                        # EMPATE
                        elif femp == True:
                            mostrarIMG8()

                    # Derecha
                    if fizq == False and fder == True:
                        # Mostramos
                        img = images[juego]
                        # Redimensionamos
                        img = imutils.resize(img, width=240, height=240)
                        ali, ani, c = img.shape
                        # Mostramos imagen
                        frame[130: 130 + ali, 30: 30 + ani] = img
                        #print(juego)

                        # Si ya jugamos
                        if fj == True and fr == False:
                            # Extraemos valores de interes
                            # Punta DI
                            x2, y2 = lista1[8][1:]
                            # Punta DC
                            x3, y3 = lista1[12][1:]
                            # Punta DA
                            x4, y4 = lista1[16][1:]

                            # Falange DI
                            x22, y22 = lista1[6][1:]
                            # Falange DC
                            x33, y33 = lista1[10][1:]
                            # Falange DA
                            x44, y44 = lista1[14][1:]

                            # Condiciones de posicion
                            # Piedra
                            if x2 > x22 and x3 > x33 and x4 > x44:
                                # IA PAPEL
                                if juego == 3:
                                    fgia,fgus,femp,fr=ganadorIA()
                                    conia=conia+1
                                # IA PIEDRA
                                elif juego == 4:
                                    fgia,fgus,femp,fr=empate()
                               #Ia Tijera
                                elif juego == 5:
                                    fgia,fgus,femp,fr=ganadorUsuario()
                                    conp1=conp1+1


                            # Papel
                            elif x2 < x22 and x3 < x33 and x4 < x44:
                                # IA PAPEL
                                if juego == 3:
                                    fgia,fgus,femp,fr=empate()
                                # IA PIEDRA
                                elif juego == 4:
                                    fgia,fgus,femp,fr=ganadorUsuario()
                                    conp1=conp1+1
                                #IA Tijera
                                elif juego == 5:
                                    fgia,fgus,femp,fr=ganadorIA()
                                    conia=conia+1

                            # Tijera
                            elif x2 < x22 and x3 < x33 and x4 > x44:
                                # IA PAPEL
                                if juego == 3:
                                    fgia,fgus,femp,fr=ganadorUsuario()
                                    conp1=conp1+1
                                     
                                # IA PIEDRA
                                elif juego == 4:
                                    fgia,fgus,femp,fr=ganadorIA()
                                    conia=conia+1
                                #IA Tijera
                                elif juego == 5:
                                    fgia,fgus,femp,fr=empate()

                        # Mostramos ganador
                        # IA
                        if fgia == True:
                            mostrarIMG6()

                        # USUARIO
                        elif fgus == True:
                            mostrarIMG7()

                        # EMPATE
                        elif femp == True:
                            mostrarIMG8()

    # 2 Jugadores
    elif jug == 2:
        # Encontramos la segunda mano
        lista2, bbox2, jug2 = detector.encontrarposicion(frame, ManoNum=1, dibujar=True, color = [255,0,0])

        # Dividimos pantalla
        cv2.line(frame, (cx, 0), (cx, 240), (255, 0, 0), 2)
        cv2.line(frame, (cx, 240), (cx, 480), (0, 255, 0), 2)

        # Mostramos jugadores
        cv2.rectangle(frame, (245, 25), (408, 60), (0, 0, 0), -1)
        cv2.putText(frame, '2 JUGADORES', (250, 50), cv2.FONT_HERSHEY_SIMPLEX, 0.71, (255, 0, 0), 2)
        
        
        #Puntaje
        cv2.putText(frame,str(contp1),(288,90),cv2.FONT_HERSHEY_SIMPLEX,1.5,(0.255,0),3)
        cv2.putText(frame,str(contp2),(330,90),cv2.FONT_HERSHEY_SIMPLEX,1.5,(0.255,0),3)

        
        #Extraemos Información importante 
        #Pulgares
        xpu1,ypu1=lista1[4][1:]#Mano 1
        xpu2,ypu2=lista2[4][1:]#Mano 2
        
        #Indice
        xin1,yin1=lista1[8][1:]
        xin2,yin2=lista2[8][1:]
        
        if ypu1<yin1 and ypu2<yin2 or fini==True:
            
            fini=True
            
            #Extraemos info
            
            xco1,yco1=lista1[9][1:]
            xco2,yco2=lista2[9][1:]
            
            #Mostramos Mensajes de Inicio
            if conteo2==0:
                #Mensajes de Inicio
                cv2.rectangle(frame,(195,425),(455,460),(0,0,0),-1)
                cv2.putText(frame, 'a la cuenta de tres',(200,450), cv2.FONT_HERSHEY_SIMPLEX,0.71,(255,0,0),2)
            #conteo
            if conteo2<=2:
                #Empezamos Conteo
                #Superamos el Umbral
                if yco1<cy-40 and yco2<cy and fd2==False:
                    fu2=True
                    cv2.circle(frame,(cx,cy),5,(255,255,0),-2)
                    
                elif yco1>cy -40 and yco2>cy -40 and fu2==True:
                    conteo2=conteo2+1
                    fu2=False
                #Mostramos
                if conteo2>=1:
                    #Mostramos Conteo
                    cv2.rectangle(frame,(290,425),(350,460),(0,0,0),-1)
                    cv2.putText(frame,str(conteo2),(306,450),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2)
            elif conteo2==3:
                #Extraemos Información que necesitamos
            
                #Jugador 1
                #Punto dedo Indice
                xpin1,ypin1=lista1[8][1:]
                #Punta dedo del Medio
                xpco1,ypco1=lista1[12][1:]
                #Punta dedo del Anular
                xpan1,ypan1=lista1[16][1:]
            
                #Falange dedo Indice
                xfin1,yfin1=lista1[6][1:]
                #Falange dedo del Medio
                xfco1,yfco1=lista1[10][1:]
                #Falange dedo del Anular
                xfan1,yfan1=lista1[14][1:]
            
                #Jugador 2
                #Punto dedo Indice
                xpin2,ypin2=lista2[8][1:]
                #Punta dedo del Medio
                xpco2,ypco2=lista2[12][1:]
                #Punta dedo del Anular
                xpan2,ypan2=lista2[16][1:]
            
                #Falange dedo Indice
                xfin2,yfin2=lista2[6][1:]
                #Falange dedo del Medio
                xfco2,yfco2=lista2[10][1:]
                #Falange dedo del Anular
                xfan2,yfan2=lista2[14][1:]
            
                #Definimos los lados 
                if xco1<cx and xco2>cx:
                    #print("lista1 izquierda y Lista 2 derecha")
                    #Condicion de posiscion mano izquierda
                    #Piedra->1
                    if xpin1<xfin1 and xpco1<xfco1 and xpan1<xfan1:
                        pizq=1
                    #Papel->2   
                    elif xpin1>xfin1 and xpco1>xfco1 and xpan1>xfan1:
                        pizq=2
                    #Tijera->3
                    elif xpin1>xfin1 and xpco1>xfco1 and xpan1<xfan1:
                        pizq=3
                
                    #Condicion de posiscion mano derecha
                    #Piedra->1
                    if xpin2>xfin2 and xpco2>xfco2 and xpan2>xfan2:
                        pder=1
                    #Papel->2   
                    elif xpin2<xfin2 and xpco2<xfco2 and xpan2<xfan2:
                        pder=2
                    #Tijera->3
                    elif xpin2<xfin2 and xpco2<xfco2 and xpan2>xfan2:
                        pder=3
                elif xco1>cx and xco2<cx:
                    #print("lista1 izquierda y Lista 2 derecha")
                    #Condicion de posiscion mano izquierda
                    #Piedra->1
                    if xpin1>xfin1 and xpco1>xfco1 and xpan1>xfan1:
                        pder=1
                    #Papel->2   
                    elif xpin1<xfin1 and xpco1<xfco1 and xpan1<xfan1:
                        pder=2
                    #Tijera->3
                    elif xpin1<xfin1 and xpco1<xfco1 and xpan1>xfan1:
                        pder=3
                
                    #Condicion de posiscion mano derecha
                    #Piedra->1
                    if xpin2<xfin2 and xpco2<xfco2 and xpan2<xfan2:
                        pizq=1
                    #Papel->2   
                    elif xpin2>xfin2 and xpco2>xfco2 and xpan2>xfan2:
                        pizq=2
                    #Tijera->3
                    elif xpin2>xfin2 and xpco2>xfco2 and xpan2<xfan2:
                        pizq=3
                    
                #print(pizq,pder)
                #Piedra Izquierda
                if pizq==1 and reset2 == False:
                    #Piedra Derecha
                    if pder==1:
                        #Mostramos
                        mostrarIMG8()
                        empate = True;
                    #Papel DErecha
                    elif pder==2:
                        #Mostramos
                        cv2.putText(frame, 'Gana Derecha', (350, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.71, (0, 255, 0), 2)
                        ganoDerecha = True;
                        if fpunt==False:
                            fpunt=True
                            contp2=contp2+1
                    #Tijera Derecha
                    elif pder==3:
                        cv2.putText(frame, 'Gana Izquierda', (150, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.71, (0, 255, 0), 2)
                        ganoIzquierda = True;
                        if fpunt==False:
                            fpunt=True
                            contp1=contp1+1
                    reset2 = True;
                #Papel Izquierda
                elif pizq==2 and reset2 == False:
                    #Piedra Derecha
                    if pder==1:
                        #Mostramos
                        cv2.putText(frame, 'Gana Izquierda', (150, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.71, (0, 255, 0), 2)
                        ganoIzquierda = True;
                        if fpunt==False:
                            fpunt=True
                            contp1=contp1+1
                    #Papel Derecha        
                    elif pder==2:
                        mostrarIMG8()
                        empate = True;
                    #Tijera Derecha
                    elif pder==3:
                        #Mostramos
                        cv2.putText(frame, 'Gana Derecha', (350, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.71, (0, 255, 0), 2)
                        ganoDerecha = True;
                        if fpunt==False:
                            fpunt=True
                            contp2=contp2+1
                    reset2 = True;
                #Tijera Izquierda 
                elif pizq==3 and reset2 == False:
                    #Piedra Derecha
                    if pder==1:
                        #Mostramos
                        cv2.putText(frame, 'Gana Derecha', (350, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.71, (0, 255, 0), 2)
                        ganoDerecha = True;
                        if fpunt==False:
                            fpunt=True
                            contp2=contp2+1
                    #Papel DErecha
                    elif pder==2:
                        #Mostramos
                        cv2.putText(frame, 'Gana Izquierda', (150, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.71, (0, 255, 0), 2)
                        ganoIzquierda = True;
                        if fpunt==False:
                            fpunt=True
                            contp1=contp1+1
                    #Tijera Izquierda
                    elif pder==3:
                        mostrarIMG8()
                        empate = True;
                    
                    reset2 = True;
                
                if(empate == True):
                    mostrarIMG8();
                elif(ganoDerecha == True):
                    cv2.putText(frame, 'Gana Derecha', (350, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.71, (0, 255, 0), 2)
                elif(ganoIzquierda == True):
                    cv2.putText(frame, 'Gana Izquierda', (150, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.71, (0, 255, 0), 2)

    # 0 Jugadores
    elif jug == 0:
        # Mostramos jugadores
        cv2.rectangle(frame, (225, 25), (388, 60), (0, 0, 0), -1)
        cv2.putText(frame, '0 JUGADORES', (230, 50), cv2.FONT_HERSHEY_SIMPLEX, 0.71, (0, 0, 255), 2)

        # Mensaje inicio
        cv2.rectangle(frame, (115, 425), (480, 460), (0, 0, 0), -1)
        cv2.putText(frame, 'LEVANTA TU MANO PARA INICIAR', (120, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.71, (0, 0, 255), 2)

    if t==82 or t==114:
        fs,fu,fd,fj,fr,fgia,fgus,femp,fder,fizq,conteo,fini,fu2,fd2,fpunt,conteo2,pizq,pder,ganoDerecha,ganoIzquierda, empate,reset2=reset()
    # Mostramos frames
    cv2.imshow("JUEGO CON AI", frame)
    if t == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
