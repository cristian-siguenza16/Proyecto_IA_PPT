```python
import cv2
import random
import SeguimientoManos as sm  # Clase manos
import os
import imutils
```

```python
# Declaracion de variables
fs = False      # Bandera de haber presionado s
fu = False      # Bandera up
fd = False      # Bandera down
fj = False      # Bandera juego
fr = False      # Bandera reset
fgia = False    # Bandera gana IA
fgus = False    # Bandera gana Usuario
femp = False    # Bandera empate
fder = False    # Bandera derecha
fizq = False    # Bandera izquierda
conteo = 0
```

```python
# Lectura de la camara
cap = cv2.VideoCapture(0)

# Declaramos el detector
detector = sm.detectormanos(Confdeteccion=0.9)
# Empezamos
while True:
    # Lectura de la videocaptura
    ret, frame = cap.read()

    # Leemos teclado
    t = cv2.waitKey(1)

    # Extraemos la mitad del frame
    al, an, c = frame.shape
    cx = int(an/2)
    cy = int(al/2)

    # Espejo
    frame = cv2.flip(frame,1)

    # Encontramos las manos
    frame = detector.encontrarmanos(frame, dibujar=True)
    # Posiciones mano 1
    lista1, bbox1, jug = detector.encontrarposicion(frame, ManoNum=0, dibujar=True, color = [0,255,0])
    # 1 Jugador
    if jug == 1:
        # Dividimos pantalla
        cv2.line(frame, (cx,0), (cx,480), (0,255,0), 2)

        # Mostramos jugadores
        cv2.rectangle(frame, (245, 25), (380, 60), (0, 0, 0), -1)
        cv2.putText(frame, '1 JUGADOR', (250, 50), cv2.FONT_HERSHEY_SIMPLEX, 0.71,(100,255,149), 2)

        # Mensaje inicio
        #cv2.rectangle(frame, (145, 425), (465, 460), (0, 0, 0), -1)
        #cv2.putText(frame, 'PRESIONA S PARA EMPEZAR', (150, 450), cv2.FONT_HERSHEY_SIMPLEX, 0.71, (0, 255, 0), 2)
        # Obtenemos posicion de la mano
        if len(lista1) != 0:
            # Extraemos las coordenadas del dedo central
            x1, y1 = lista1[9][1:]
            # Definimos en que parte de la pantalla esta
            # Izquierda
            if x1 < cx:
                # Banderas Lado
                fizq = True
                fder = False
                # Derecha
            elif x1 > cx:
                # Banderas Lado
                fder = True
                fizq = False    
        if fizq == True and fder == False:
            # Extraemos valores de interes
        
            # Punta Dedo Indice
            x2, y2 = lista1[8][1:]
            # Punta Dedo Centro 
            x3, y3 = lista1[12][1:]
            # Punta Dedo anular
            x4, y4 = lista1[16][1:]

            # Falange Dedo Indice
            x22, y22 = lista1[6][1:]
            # Falange Dedo Centro
            x33, y33 = lista1[10][1:]
            # Falange Dedo anular
            x44, y44 = lista1[14][1:]

            # Condiciones de posicion
            # Piedra
            if x2 < x22 and x3 < x33 and x4 < x44:
                print("Piedra",'/n')
                cv2.putText(frame, 'PIEDRA', (125, 460), cv2.FONT_HERSHEY_SIMPLEX, 0.71,(100,255,149), 2)
            #Papel    
            elif x2 > x22 and x3 > x33 and x4 > x44:
                print("Papel",'/n')
                cv2.putText(frame, 'PAPEL', (125, 460), cv2.FONT_HERSHEY_SIMPLEX, 0.71,(100,255,149), 2)
            # Tijera
            elif x2 > x22 and x3 > x33 and x4 < x44:
                print("Tijera",'/n')
                cv2.putText(frame, 'TIJERAS', (125, 460), cv2.FONT_HERSHEY_SIMPLEX, 0.71,(100,255,149), 2)

        if fizq == False and fder == True:
            # Extraemos valores de interes
            # Punta DI
            x2, y2 = lista1[8][1:]
            # Punta DC
            x3, y3 = lista1[12][1:]
            # Punta DA
            x4, y4 = lista1[16][1:]

            # Falange DI
            x22, y22 = lista1[6][1:]
            # Falange DC
            x33, y33 = lista1[10][1:]
            # Falange DA
            x44, y44 = lista1[14][1:]

            # Condiciones de posicion
            # Piedra
            if x2 > x22 and x3 > x33 and x4 > x44:
                print("Piedra",'/n')
                cv2.putText(frame, 'PIEDRA', (375, 460), cv2.FONT_HERSHEY_SIMPLEX, 0.71,(100,255,149), 2)
            # Papel    
            elif x2 < x22 and x3 < x33 and x4 < x44:
                print("Papel",'/n')
                cv2.putText(frame, 'PAPEL', (375, 460), cv2.FONT_HERSHEY_SIMPLEX, 0.71,(100,255,149), 2)
            # Tijeras    
            elif x2 < x22 and x3 < x33 and x4 > x44:
                print("Tijeras",'/n')
                cv2.putText(frame, 'TIJERAS', (375, 460), cv2.FONT_HERSHEY_SIMPLEX, 0.71,(100,255,149), 2)
    cv2.imshow("JUEGO CON AI", frame)
cap.release()
cv2.destroyAllWindows()
```
